```python
import os
import pandas as pd
```

```python
data = []
for fil in os.listdir()[1:]:
    if fil.endswith('.csv'):
        data.append(fil)
data 
```

```python
df = pd.DataFrame()
category = ['Girl Clothing','Watches','Boy Clothing','Boy Clothing','Boy Clothing','Men Clothing','Women Clothing','Girl Clothing','Girl Clothing']
for datafile, cate in zip(data, category):
    df_file = pd.read_csv(datafile, index_col=[0])
    df_file.insert(4, column='Category', value = cate)
    df = pd.concat([df,df_file])
df 
```

```python
df.to_csv('AmazonData.csv', index=False)
```
